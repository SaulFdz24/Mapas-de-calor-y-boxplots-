```python
import pandas as pd
import numpy as np

fact = pd.read_excel('Info Reto.xlsx',sheet_name='facturacion')
gastos = pd.read_excel('Info Reto.xlsx',sheet_name='gastos y costos')
productos = pd.read_excel('Info Reto.xlsx',sheet_name='productos2022')
saldos = pd.read_excel('Info Reto.xlsx',sheet_name='saldos2023')
notas = pd.read_excel('Info Reto.xlsx',sheet_name='notas credito')
devol = pd.read_excel('Info Reto.xlsx',sheet_name='devoluciones')
```

```python
dataframes = [fact, gastos, productos, saldos, notas, devol]

for i, df in enumerate(dataframes):
    dataframes[i] = df.fillna(method='bfill')
    display(dataframes[i].info())
```

##Histogramas

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
sns.set_style("darkgrid")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.histplot(data=fact, x='CAN_TOT', kde=True, ax=axes[0])
sns.histplot(data=fact, x='DES_TOT', kde=True, ax=axes[1])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

sns.histplot(data=gastos, x='TOTAL_MX', kde=True, ax=axes[0,0])
sns.histplot(data=gastos, x='IVA', kde=True, ax=axes[0,1])
sns.histplot(data=gastos, x='TOTAL_SAT', kde=True, ax=axes[1,0])
sns.histplot(data=gastos, x='IMPORTE', kde=True, ax=axes[1,1])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

sns.histplot(data=productos, x='PRECIO_UNITARIO', kde=True, ax=axes[0,0])
sns.histplot(data=productos, x='COSTO_UNITARIO_CALCULADO', kde=True, ax=axes[0,1])
sns.histplot(data=productos, x='MARGEN_UNITARIO_CALCULADO', kde=True, ax=axes[1,0])
fig.delaxes(axes[1, 1])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.histplot(data=saldos, x='PLAZO', kde=True, ax=axes[0])
sns.histplot(data=saldos, x='MONTO ADEUDADO', kde=True, ax=axes[1])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.histplot(data=notas, x='CAN_TOT', kde=True, ax=axes[0])
sns.histplot(data=notas, x='CVE_VEND', kde=True, ax=axes[1])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.histplot(data=devol, x='CAN_TOT', kde=True, ax=axes[0])
sns.histplot(data=devol, x='CVE_VEND', kde=True, ax=axes[1])
```

##Mapas de calor

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(30, 30))

sns.heatmap(data=fact.drop('Unnamed: 0', axis=1).corr(), annot=True, cmap='seismic', linewidth=1, ax=axes[0, 0])
axes[0, 0].set_title('FACTURACIÓN', fontsize=30)

sns.heatmap(data=gastos.drop('Unnamed: 0', axis=1).corr(), annot=True, cmap='seismic', linewidth=1, ax=axes[0, 1])
axes[0, 1].set_title('GASTOS Y COSTOS', fontsize=30)

sns.heatmap(data=productos.drop('Unnamed: 0', axis=1).corr(), annot=True, cmap='seismic', linewidth=1, ax=axes[1, 0])
axes[1, 0].set_title('PRODUCTOS', fontsize=30)

sns.heatmap(data=saldos.drop('Unnamed: 0', axis=1).corr(), annot=True, cmap='seismic', linewidth=1, ax=axes[1, 1])
axes[1, 1].set_title('SALDOS', fontsize=30)

sns.heatmap(data=notas.drop('Unnamed: 0', axis=1).corr(), annot=True, cmap='seismic', linewidth=1, ax=axes[2, 0])
axes[2, 0].set_title('NOTAS DE CRÉDITO', fontsize=30)

sns.heatmap(data=devol.drop('Unnamed: 0', axis=1).corr(), annot=True, cmap='seismic', linewidth=1, ax=axes[2, 1])
axes[2, 1].set_title('DEVOLUCIONES', fontsize=30)

plt.tight_layout()

plt.show()
```
